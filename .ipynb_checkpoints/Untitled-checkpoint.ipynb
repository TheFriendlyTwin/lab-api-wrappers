```python
import numpy as np
import pandas as pd

# Importing the track data csv file
track_data = pd.read_csv(r"C:\Users\mafal\Documents\ironhack\labs\lab-unsupervised-learning-intro\files_for_lab\spotify_track_data.csv")
track_data
```

```python
import ast  # Import the Abstract Syntax Trees module

# Function that extracts the artist name from the column album.artist
def extract_artist_name(artist_album_str):
    artist_album_list = ast.literal_eval(artist_album_str)
    artist_name = artist_album_list[0]['name']
    return artist_name

# Adding artist_name column to our track data df
track_data['artist_name'] = track_data['album.artists'].apply(extract_artist_name)
track_data[['id', 'name', 'album.artists', 'artist_name']]
```

```python
# Function that extracts the artist id from the column album.artist
def extract_artist_id(artist_album_str):
    artist_album_list = ast.literal_eval(artist_album_str)
    artist_id = artist_album_list[0]['id']
    return artist_id

# Adding artist_id column to our track data df
track_data['artist_id'] = track_data['album.artists'].apply(extract_artist_id)
track_data[['id', 'name', 'album.artists', 'artist_name', 'artist_id']]
```

```python
# Save the track_data df to a CSV file
track_data.to_csv(r"C:\Users\mafal\Documents\ironhack\labs\lab-unsupervised-learning-intro\spotify_track_data.csv", index=False)
```

```python
from collections import defaultdict

# Function to get genres for a list of artists using batch requests
def get_genres_for_artist_ids(artist_ids):
    genres_dict = defaultdict(list)
    unique_artist_ids = list(set(artist_ids))  # Remove duplicates to minimize requests
    batch_size = 50  # Spotify allows fetching up to 50 artists in one request
    
    for i in range(0, len(unique_artist_ids), batch_size):
        batch = unique_artist_ids[i:i+batch_size]
        try:
            artists_info = sp.artists(batch)
            for artist in artists_info['artists']:
                # Taking the first genre for simplicity, you could customize this part
                genres_dict[artist['id']] = artist['genres'][0] if artist['genres'] else 'Unknown'
        except Exception as e:
            print(f"Error fetching batch {i}: {e}")
    
    return genres_dict

# Applying the function to the entire DataFrame's artist IDs
# Note: It might be more efficient in terms of memory and speed to directly use the series here
artist_ids = track_data['artist_id'].tolist()
genres_dict = get_genres_for_artist_ids(artist_ids)

# Map the dictionary back to the DataFrame
track_data['artist_genre'] = track_data['artist_id'].map(genres_dict)
```
